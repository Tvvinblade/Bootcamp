## Работа с правами доступа

Давайте посмотрим на вывод следующей команды. Нас интерисует начало

```python
!chmod 777 file.txt
!ls -l ./file.txt
```

# Что означает `-rwxrwxrwx`

- Первый символ — тип файла:
  - `-` — обычный файл
  - `d` — директория (папка)
  - `l` — символическая ссылка (по сути это когда вы ссылаететь на файл из другого каталога)

- Следующие 9 символов разделены на три группы по три символа каждая:

| Позиция        | Символы | Значение                           |
|----------------|---------|----------------------------------|
| 1–3 (владелец) | `rwx`   | Чтение (r), запись (w), выполнение (x) для владельца файла |
| 4–6 (группа)   | `rwx`   | Те же права для группы пользователей файла                  |
| 7–9 (остальные)| `rwx`   | Те же права для всех остальных пользователей                 |

`chmod -x script.sh` — убираем право на исполнение файла

```python
!chmod -x file.txt
!ls -l ./file.txt
```

`chmod +x script.sh` — убираем право на исполнение файла

```python
!chmod +x file.txt
!ls -l ./file.txt
```

`chmod 644 file.txt` — установить права доступа

```python
!chmod 764 file.txt
!ls -l ./file.txt
```
