## Пробрасывание SSH-ключа в GITHUB

### Генерация ключа

````python
ssh-keygen # пропускаем 3 последующих команды через Enter
````

### Копируем `/root/.ssh/id_rsa.pub` и переходим в GitHub и вставляем данный публичный ключ 

````python
cat /root/.ssh/id_rsa.pub
````

### Необходимо создать репозиторий в GITHUB

### Создаем и клонируем репозиторий(при первом клонировании спросит о подтверждении хоста github, как доверительно)

````python
git clone git@github.com:ShustGF/lesson_git.git
cd ./lesson_git/
````

### **Добавляем** новый файл, **изменяем** файл и **удаляем** файл. Смотрим как меняется статус

````python
git status
ls -la # показать скрытые файлы в которых лежат отслеживаемые файлы
echo 'Строка 1' >> file.txt
echo 'какой-то текст' >> README.md
rm README.md
````

### Добавить данные на **индексирование** и создаем **первый коммит**

````python
git add .
git status # видим что изменения зафиксировались
git commit -m "first commit" # бах оибочка вышла
````

При первом коммите, на новом компьютере, нужно будет прописать кто является пользователем по умолчанию для коммитов
!!! Сам Git подскажет вам команду


````python
git config --global user.email "you@example.com" && git config --global user.name "Your Name"
````

Еще раз коммитим

````python
git commit -m "first commit"
````

Необходимо свои изменения запушить(загрузить) в ГИТ

````python
git push
````

Теперь самое сложное, разберемся в ветках. Создадим первую ветку

````python
git branch shustikov/branch-1 # Создаем
git branch
git checkout shustikov/branch-1 # переходим на новую ветку
git branch
````

Изменяем файл (добавь 3 строки)

git add . && git commit -m "add new strings"
git push 

````python
git push --set-upstream origin shustikov/branch-1
````

Возращаемся на мастер ветку и **скачиваем** все изменения проекта

````python
git pull
````

Посмотрим на коализию веток

перед созданием новой ветки покажи утилиту `GitLens — Git supercharged`

````python
git checkout -b shustikov/branch-2
````

Внести изменения в файл

````python
git add . && git commit -m "commit branch 2" 
git push origin shustikov/branch-2
````

Возвращается на ветку shustikov/branch-1 и меняем тот же файл
